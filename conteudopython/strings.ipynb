```python
# Concatenação, junção de strings p/ formação de outra string
nome1 = "Alice"
idade1 = 30
concatenacao = "Meu nome é " + nome1 + " e tenho " + str(idade1) + " anos."
print(concatenacao)

# o "+" permite juntarmos variáveis sem espaços
a = "Joso Leirbag "
b = 16
c = "Chama"
d = "1"*20
frase1 = a + str(b) + c + d + "!"
print(f"Juntando tudo: '{frase1}'.")
```

```python
# Interpolação de strings, método .format ou f-strings

# .format, introdução de variáveis dentro de uma string
nome1 = "João"
idade1 = 25
frase1 = "Olá, meu nome é {} e tenho {} anos."
format_frase = frase1.format(nome1, idade1)
print(format_frase)

# índices, com eles podemos determinar a posição da variável que vai entrar nas respectivas chaves
frase2 = "Olá, meu nome é {1} e tenho {0} anos.".format(nome1, idade1)
print(frase2)

print()

# Formatação dentro do .format

# número de casas decimais de um dado número
numero_inteiro = 42
numero_ponto_flutuante = 5.1234
frase3 = "Número inteiro: {:d}, Número de ponto flutuante {:.2f}".format(numero_inteiro, numero_ponto_flutuante)
print(frase3)

# número de espaços à direita e sequerda, respectivamente
nome2 = "Alice"
idade2 = 30
frase = "Nome: {:<10} | Idade: {:>5}".format(nome2, idade2)
print(frase)

print()

# F-strings, mesmo do .format porém inserimos o nome da variável diretamente nas chaves
nome3 = "Luis"
idade3 = 30
frase4 = f"Olá, meu nome é {nome3} e tenho {idade3} anos."
print(frase4)

# assim como no .format, podemos inserir expressões dentor das chaves
valor = 42
frase5 = f'O dobro de {valor} é {valor * 2}, e o triplo é {valor * 3}.'
print(frase5)

# também podemos formatar nosso texto
nome4 = "Ana"
idade4 = 25
altura = 1.68
frase6 = f"Nome: {nome4}, Idade: {idade4:d}, Altura: {altura:.2f}m"
print(frase6)
```

```python
# Manipulação de strings

texto1 = "TExTo com minúsculas e MAIÚSCUlas."
print(f"Texto original: {texto1}")

print()

# frase[m:n:p], fatiamento da string de m até (n-1) passo p
print(texto1[2:9])
print(texto1[3:16:2])
#vai de 3 até 11 pulando de 2 em 2
print(texto1[:15])
print(texto1[15:])
#do início até o 14; do 15 até o fim, também funcionam com salto

# .lower(), deixa a string toda em minúsculas
minus_texto1 = texto1.lower()
print(f"Texto em minúsculas: {minus_texto1}")

# .upper(), deixa a string toda em maiúsculas
maius_texto1 = texto1.upper()
print(f"Texto em maiúsculas: {maius_texto1}")

# .islower(), verifica se uma string está em minúsculas
# .isupper(), verifica se uma string está em maiúsculas
# invertendo a 'maiuscularidade' de uma frase
print("Texto com tamanhos de letras invertidos: ", end = "")
for x in texto1:
  if x.islower():
    print(x.upper(), end = "")
  else:
    print(x.lower(), end = "")
print()

# .capitalize(), deixa apenas o primeiro caractere em maiúscula, o restante em minúsculas
capit_texto1 = texto1.capitalize()
print(f"Texto formatado: {capit_texto1}")

# .title(), deixa somente a primeira letra de cada palavre em maiúscula no texto
titulo_texto1 = texto1.title()
print(f"Texto titularizado: {titulo_texto1}")

print()

# .count(), conta quantas vezes uma certa string aparece dentro de outra
palavra_chave = "TExTo"
count_palavra_chave_texto1 = texto1.count(palavra_chave)
print(f"'{palavra_chave}' aparece {count_palavra_chave_texto1} vezes no texto1.")

texto_teste = "Maria foi ao mercado comprar abóbora."
count_a_texto_teste = texto_teste.count("a")
print(f"A letra 'a' aparece {count_a_texto_teste} vezes no texto teste.")

print(f"'a' aparece {texto1.count('a', 1, 13)} vezes no texto1 de 1 até 12.")

# .find(), diz em que posição começa uma string dentro do texto
print(f"'com' começa no texto1 na posição de índice {texto1.find('com')}.")
print(f"A primeira aparição da letra 'a' no texto1 foi no índice {texto1.find('a')} e a última aparição foi no índice {texto1.rfind('a')}.")
print(f"Como 'chama' não existe no texto1 o índice de 'chama' é: {texto1.find('chama')}.")
#retorna -1 qunado a string não existe dentro texto

# .replace(), substitui uma string no lugar de outra
subst_texto1 = texto1.replace("com minúsculas e MAIÚSCUlas.", "1")
print(f"Substituindo strings: {subst_texto1}")

print()

# .split(), quebra a string em "n+1" pedaços onde "n" é a quantidade de vezes que um certo separador aparece na string
frag1, frag2, frag3, frag4, frag5 = texto1.split(" ")
print(f"Palavras do texto1: '{frag1}', '{frag2}', '{frag3}', '{frag4}', '{frag5}'")

print()

# .strip(), tira os espaços iniciais e finais de uma string
texto2 = "  Meu nome é José .   "
print(f"Texto2: '{texto2}'.")
print(f"Texto2 sem espaços inúteis: '{texto2.strip()}'.")
print(f"Texto2 sem espaços inúteis à esquerda: '{texto2.lstrip()}'.")
print(f"Texto2 sem espaços inúteis à direita: '{texto2.rstrip()}'.")

print()

# ''.join(), coloca uma string entre cada caractere de outra string
print("-".join(texto1))
print("-".join(texto2.split()))

print()

# .isdigit(), retorna True se a string for somente dígitos e False se não for somente de dígitos
digit_texto1 = texto1.isdigit()
if digit_texto1:
  print("O texto1 é feito somente de núemros.")
else:
  print("O texto1 não é feito somente de dígitos.")
print(f"O .isdigit() é do tipo '{type(digit_texto1)}'")

# .isalpha(), retorna True se a string for feita somente de letras e se não tiver vazia e retorna False se não for somente de letras ou se for vazia
alpha_texto1 = texto1.isalpha() #Verificar
if alpha_texto1:
  print("O texto1 é formado somente com letras")
else:
  print("O texto1 não é formado apenas por letras.")

print()

# .partition("separador"), quebra uma string em uma tupla com 3 sub-strings
# retorna uma tupla contento o separador, oq vem antes dele e o que vem depois dele
print("a-cs-s".partition("-"))
# caso o separador não seja encontrado, retorna uma tupla com a string completa e duas strings vazias
print("a-cs-s".partition("_"))
```

```python
# identificações de um nome

nome_completo = "João Gabriel da Silva Sousa"
print(nome_completo.lower())
print(nome_completo.upper())
print("Seu nome tem")
print(len(nome_completo.replace(" ", '')))
print("letras.")
print(f"Seu primeiro nome é '{nome_completo.split()[0]}' com {len(nome_completo.split()[0])} letras.")
print(f"Seu último é '{nome_completo.split()[len(nome_completo.split())-1]}'.")
if 'silva' in nome_completo.lower():
  print("Seu nome tem 'Silva'.")
else:
  print("Seu nome não tem 'Silva'.")

print(f"A posição da primeira e última letras 'a' são, respectivamente, {nome_completo.find('a')-nome_completo[:nome_completo.find('a')+1].count(" ") + 1} e {nome_completo.rfind('a')-nome_completo[:nome_completo.rfind("a")].count(" ")+1}.")
```

```python
# identificando se uma frase é palíndroma, palavras sem acento

frase = input("Digite uma frase: ").strip().lower()
palavras = frase.split()
frase_junto = ''.join(palavras) #só funciona se os elementos da lista frase.split() forem str
frase_reversa = ''

for x in range(len(frase_junto)-1, -1, -1):
  frase_reversa += frase_junto[x]
if frase_junto == frase_reversa:
  print("Sua frase é palíndroma.")
else:
  print("Sua frase não é palíndroma.")
```
