```python
# Funções, rotinas

# começamos com o comando def, em seguida vem o nome da função e seus parâmetros
def produto(a, b):
    print(a*b)
# a partir daqui, podemos "chamar" a função quantas vezes quizermos desde que haja concordância entre os parâmetros dados

produto(4, 5)
produto(1, 2)
produto(7, 0)
# produto(1) => Error, pois há poucos parâmetros para esta função
# produto(1, 2, 3) => Error, pois há muitos parâmetros, esta função só trabalho com dois

print()

# também podemos mudar a ordem dos parâmetros na "chamada" da função
def diferença(a, b):
    print(f"{a} - {b} = {a-b}")

diferença(10, 4)
diferença(a = 5, b = 1)
diferença(b = 5, a = 1)

print()

# generalizando parâmetros
# o comando *valores diz que o usuário vai colocar uma quantidade desconhecida de valores neste parâmetro, evitando erros anteriores
def soma(*valores):
    soma = 0
    for x in valores:
        soma += x
    print(soma)

soma(1, 2)
soma(7)
soma()

print()

# os parâmetros não têm limites de tipagem, desde que o usuário utilize da forma correta
def produto_escalar(a, lista):
    for x in range(0, len(lista)):
        lista[x] *= a

a = [1, 2, 3]
produto_escalar(3, a)
print(f"a * 3 = {a}")

print()

# também podemos misturar entre parâmetros comuns e generalizados
# o Python cria uma tupla para este, caso tenha len() > 1
def chama(a, *valores):
    for x in valores:
        print(a*x, end = " ")
    print()

chama(2, 1, 2, 3)
chama(0, 1)
chama(10, 6, 3, 8, 10)

print()

# geralmente se utiliza o comando return, para assim, somente termos o valor em mãos e não necessariamente imprimí-lo
def maior(lista):
    maior = lista[0]
    for x in lista:
        if x > maior:
            maior = x
    return maior

b = [3, 8, 4, 3, 1, 7]
maior_b = maior(b)
print(f"O maior elemento de b é {maior_b}.")

# parâmetros opcionais, caso não seja informado um valor ao parâmetro, ele recebe seu valor padrão dito pelo programador
def soma_opc(a, b, c = 0):
    # caso não seja dado um valor para c, este recebe zero
    soma = a+b+c
    print(f"A soma vale {soma}.")

soma_opc(1, 2, 3)
soma_opc(4, 5)

# nada impede de todos serem opcionais
def soma_opcional(a = 0, b = 0, c = 0):
    s = a + b + c
    print(f"A soma entre {a}, {b}, e {c} vale {s}.")

soma_opcional(6)

print()

# também podemos informar nossa própria explicação sobre nossas funções
# basta colocar entre aspas triplas no início da definição da função
def escreva(texto):
    '''
    --> Função que imprime sua mensagem formatada por linha
    :param texto: Mensagem que você quer que seja formatada
    EX.: escreva("Bom dia")
       -----------
         Bom dia
       -----------
    --> Função criada por joso_leirbag em meio estudo sobre Python
    '''
    tamanho = len(texto) + 4
    print("-"*tamanho)
    print(f"  {texto}")
    print("-"*tamanho)

escreva("Python é massa")
help(escreva)
```

```python
# Tratamento de erros, Exceptions

try:        # "Tente fazer":
    a = int(input("Número: "))
    r = 5 / a
except (ValueError, TypeError):     # Caso dê erro em valor ou tipagem de variável:
    print("Tivemos um problema com os tipos de dados que você digitou.")
except ZeroDivisionError:
    print("Não é possível dividir por zero!")
except KeyboardInterrupt:
    print("O usuário preferiu não informar os dados!")
except Exception as erro:
    print(f"O erro encontrado foi {erro.__cause__}")
else:       # Se não der erro:
    print(f"5 / {a} = {r:.2f}")
finally:    # Independente do que aconteça, faça:
    print("Tenha um bom dia.")
```
