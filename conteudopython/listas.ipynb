```python
# Listas, conjunto de elementos sem restrição de tipagem de variável

lista1 = [1, 2, 3, 4, 5]
print(lista1)
for x in lista1:
  print(x)

print(lista1[1:4:2])
#lista que vai de [1] até [4-1] (==[3]) pulando de 2 em 2

print()

# tamanho da lista, comando len(lista)
lista2 = ["chama1", "chama2", "chama3", "chama4", "chama5"]
tam_lista2 = len(lista2)
print(f"A lista2 possui {tam_lista2} elementos.")

len_hw = len("Hello world!")
print(f"'Hello world!' possui {len_hw} caracteres.")

# podemos criar listas usanod o método list(), por exemplo:
lista3 = list(range(2, 10, 2))
print(f"Lista3: {lista3}")

# .sort() e .sort(reverse = True), organizando uma lista p/ ordem crescente e decrescente ou ordem alfabética
# não suporta listas alfanuméricas, ou há somente números (int ou float) ou há somente letras (str)
lista4 = [1, 67, 5, 4, 9, 10.54, 78, 10.57]
lista4.sort()
print(f"Ordem crescente: {lista4}")
lista4.sort(reverse = True)
print(f"Ordem decrescente: {lista4}")

# Organização das listas, índices: lista[índice]
i1 = 0
# i1 < tam_lista2 é suficiente p/ passo 1
while i1 <= (tam_lista2 - 1):
  print(f"Elemento de índice {i1}: {lista2[i1]}")
  i1 = i1 + 1

print()

# .append(), adicionando elementos ao final de uma lista
lista2.append("chama6")
print(lista2)

# .insert(índice, valor), adicionando elementos em índices quaisquer da lista
lista2.insert(4, "Chama")
print(lista2)

# .extend(), união de listas
lista1.extend(lista2)
print(lista1)

# del, deletando elementos de uma lista pelo índice
del lista1[10]
print(lista1)

# .remove(), deletando elementos de uma lista pelo conteúdo
# elimina o primeiro elemento com o valor parametrado
lista2.remove("Chama")
print(lista2)

# .pop(), elimina o última elemento de uma lista
# também pode funcionar como del, por exemplo, lista.pop(3) == del lista[3]
lista2.pop()
print(lista2)

i2 = 5
while i2 < len(lista1):
  del lista1[i2]

print(lista1)

print()

# o jeito certo de criar cópias de listas
a = [1, 2, 3, 4, 5]
b = a      # desta forma, as duas listas estão interligadas, tudo o que acontecer com uma vai acontecer com a outra
c = a[:]   # assim, c recebe os itens de a
print(f"Lista A antes: {a}")
b[2] = 5
print(f"Lista A depois: {a}")
print(f"Lista B: {b}")
print(f"Lista C: {c}")
```

```python
# podemos somar os números que estão em uma lista
lista3 = [1, 2, 3, "chama1", 4, 5, "chama2"]
lista4 = []
for x in lista3:
  if type(x) == int:
    lista4.append(x)

soma_lista3 = sum(lista4)
print(f"A soma dos números da lista é {soma_lista3}.")

# brincando com os números dentro da lista
for item in lista3:
  if type(item) == int:
    if item % 2 == 1:
      print("Ímpar:", item)
```
